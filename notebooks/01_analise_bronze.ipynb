## 01. Análise da Tabela de Vendas

```python
df_vendas = pd.read_csv(r'01_bronze\vendas_2023_2024.csv')
print(f'{df_vendas.head(5)} \n')

# diagnositco inicial da tabela vendas
print(f'{df_vendas.info()} \n')
print(f'{df_vendas.shape} \n')

# verificando se existem valores duplicados
duplicadas_vendas = df_vendas[df_vendas.duplicated(keep=False)]
print(duplicadas_vendas)
# neste caso, nenhuma coluna id existe dados repetidos


print(df_vendas.head(5))
```
