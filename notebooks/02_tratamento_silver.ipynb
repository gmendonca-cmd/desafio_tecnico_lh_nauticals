# Importando Bibliotecas

```python
import pandas as pd
```

# Transformando Tabelas

## 01. Tabela Vendas

```python
df_vendas = pd.read_csv(r'01_bronze\vendas_2023_2024.csv')

# transormando a coluna de data de object para o tipo datatime
df_vendas['sale_date'] = pd.to_datetime(df_vendas['sale_date'], format='mixed', dayfirst=True)
print(df_vendas.info())

print(df_vendas.head(5))
```

## 02 Tabela Produtos

```python
df_produtos = pd.read_csv(r'01_bronze\produtos_raw.csv')
# print(df_produtos.head(5))

# alterando o tipo da coluna price para int
df_produtos['price'] = df_produtos['price'].str.replace(r'\D+', '', regex=True).astype(float)

print(df_produtos.info())

# padrozinadndo dados da coluna categoria 
df_produtos['actual_category'] = df_produtos['actual_category'].str.strip().str.lower()

# fazendo a padrozinação dos categorias fora do padrao
mapa_categorias = {
    'eletrônicos': 'Eletrônicos',
    'eletronicos': 'Eletrônicos',
    'eletrunicos': 'Eletrônicos',
    'eletronicoz': 'Eletrônicos',
    'eletroniscos': 'Eletrônicos',
    'e l e t r ô n i c o s': 'Eletrônicos',
    
    'propulsão': 'Propulsão',
    'propulsao': 'Propulsão',
    'propulssão': 'Propulsão',
    'propução': 'Propulsão',
    'propulçao': 'Propulsão',
    'propulsam': 'Propulsão',
    'propulção': 'Propulsão',
    'prop': 'Propulsão',
    'p r o p u l s ã o': 'Propulsão',
    
    'ancoragem': 'Ancoragem',
    'ancoragen': 'Ancoragem',
    'encoragem': 'Ancoragem',
    'ancoraguem': 'Ancoragem',
    'ancorajm': 'Ancoragem',
    'ancorajem': 'Ancoragem',
    'ancorajen': 'Ancoragem',
    'encoragi': 'Ancoragem',
    'a n c o r a g e m': 'Ancoragem'
}



# substituindo os dados padronizados
df_produtos['actual_category'] = df_produtos['actual_category'].replace(mapa_categorias)
print(f'{df_produtos.head(5)}\n')

# verificando
categorias_unicas = df_produtos['actual_category'].unique()
print(categorias_unicas)
```

## 03. Tabela Custos

```python
df_custos = pd.read_json(r'01_bronze\custos_importacao.json')

# transformadno a coluna de historico de data para lista em linhas)
df_custos_exploded = df_custos.explode('historic_data')

# tranformando os dicionários em colunas
df_historico = df_custos_exploded['historic_data'].apply(pd.Series)
print(df_historico.head(5))

#juntando tudo novamente
df_final = pd.concat([df_custos_exploded.drop(columns='historic_data'), df_historico], axis=1)

# convertendo a coluna de data de object para datetime
df_final['start_date'] = pd.to_datetime(df_final['start_date'], dayfirst=True)

print(df_final.info())
```
